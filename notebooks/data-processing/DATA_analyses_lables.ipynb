```python
import os

import pandas as pd
```

We want to know:

* The distribution of labels per labeler overall
* Gather 10 "clean" examples where the labelers diverge by 1 (negative - positive)

## Distribution of labels per labeler per decade

### Marin General distribution

```python
DECADES = ["1970s", "1980s", "1990s"]
TYPES = ["coal", "gas", "oil"]
LABELER = "marin"

dists_marin = []
for DECADE in DECADES:
    for TYPE in TYPES:
        name = f"{LABELER}_{DECADE}_{TYPE}_cleaned.csv"
        df = pd.read_csv(os.path.join(DATA_DIR, name))
        dist_labels = df.labels.value_counts()
        dists_marin.append(dist_labels)
```

```python
sum(dists_marin)
```

### Edo General distribution

```python
DECADES = ["1970s", "1980s", "1990s"]
TYPES = ["coal", "gas", "oil"]
LABELER = "edo"

dists_edo = []
for DECADE in DECADES:
    for TYPE in TYPES:
        name = f"{LABELER}_{DECADE}_{TYPE}_cleaned.csv"
        df = pd.read_csv(os.path.join(DATA_DIR, name))
        dist_labels = df.labels.value_counts()
        dists_edo.append(dist_labels)
```

```python
sum(dists_edo)
```
